```python
import zipfile
import xml.etree.ElementTree as ET

def extract_field_codes(docx_path):
    with zipfile.ZipFile(docx_path, 'r') as docx_zip:
        with docx_zip.open('word/document.xml') as document_xml:
            tree = ET.parse(document_xml)
            root = tree.getroot()
            namespaces = {'w': 'http://schemas.openxmlformats.org/wordprocessingml/2006/main'}
            field_codes = root.findall('.//w:fldSimple', namespaces)
            print(f"Found {len(field_codes)} field codes in the document.")
            for field_code in field_codes:
                print(ET.tostring(field_code, encoding='unicode'))

# Usage example with the existing file_path variable
extract_field_codes(r"NYLB\doc_NYLB 32 (revision copy).docx")
```

```python
from docx import Document

def read_docx(docx_file):
    # Open the .docx file
    doc = Document(docx_file)
    
    # Initialize a list to store all paragraphs
    all_paragraphs = []
    
    # Iterate over each paragraph in the document
    for para in doc.paragraphs:
        # Append the text of the paragraph to the list
        all_paragraphs.append(para.text)
    
    # Join the list into a single string, separating paragraphs with newlines
    document_text = "\n".join(all_paragraphs)
    
    return document_text

# Replace 'your_document.docx' with the path to your .docx file
document_text = read_docx('NYLB/NYLB 32 (revision copy).docx')
print(document_text)
```
